```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
# Cargar datasets
data_x, data_y = datasets.load_diabetes(return_X_y=True)

# Una característica 
data_x = data_x[:, np.newaxis, 8]
print('Dataset: ', data_x.shape)
```

```python
# Conjunto de pruebas 20%
x_train, x_test, y_train, y_test = train_test_split(data_x, data_y, test_size=0.2, random_state=0)
# x_train, x_test, y_train, y_test = data_x[-20:], data_x[:-20], data_y[-20:], data_y[:-20]
print(f"X_train: {x_train.shape}")
print(f"X_test: {x_test.shape}")
print(f"Y_train: {y_train.shape}")
print(f"Y_test: {y_test.shape}")
```

```python
# Modelo: Regresión lineal
model = LinearRegression()
# Entrenar el modelo
model.fit(x_train, y_train)
```

```python
ax = plt.axes()
ax.scatter(x_train, y_train, color='blue', label='Datos de entrenamiento')
ax.scatter(x_test, y_test, color='red', label='Datos de prueba')
```

```python
# Predicciones con datos de prueba
y_pred = model.predict(x_test)
print('Y_pred', y_pred.shape)
```

```python
# Evaluaciones
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print('MSE: ', mse)
print('R^2: ', r2)
```

```python
print(f"Pendiente / Coeficiente: {model.coef_}")
print(f"Ordenada de origen / Intercepto: {model.intercept_}")
```
